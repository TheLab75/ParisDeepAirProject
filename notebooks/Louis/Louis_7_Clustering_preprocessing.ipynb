# Clustering

```python
cluster1_O = ['PA75007_1', 'PA75007_2', 'PA75016', 'PA92220', 'PA92800' ]
cluster2_N = ['PA75018', 'PA93300', 'PA93500']
cluster3_E = ['PA75012_1', 'PA75012_2', 'PA75012_3']
cluster4_S = ['PA75013', 'PA75014', 'PA75015']
cluster5_C = ['PA75001', 'PA75002', 'PA75004', 'PA75006', 'PA75008', 'PA75009']
```

## Cluster 1 - Ouest

```python
df_1 = pd.read_csv(f"../../data/pollution/2_Processed/{cluster1_O[0]}.csv").copy()
df_2 = pd.read_csv(f"../../data/pollution/2_Processed/{cluster1_O[1]}.csv").copy()
df_3 = pd.read_csv(f"../../data/pollution/2_Processed/{cluster1_O[2]}.csv").copy()
df_4 = pd.read_csv(f"../../data/pollution/2_Processed/{cluster1_O[3]}.csv").copy()
df_5 = pd.read_csv(f"../../data/pollution/2_Processed/{cluster1_O[4]}.csv").copy()
```

```python
df_1 = df_1.drop(columns=['Station_name', 'Station_type'])
df_2 = df_2.drop(columns=['Station_name', 'Station_type'])
df_3 = df_3.drop(columns=['Station_name', 'Station_type'])
df_4 = df_4.drop(columns=['Station_name', 'Station_type'])
df_5 = df_5.drop(columns=['Station_name', 'Station_type'])
```

```python
df_merge = pd.merge(df_1, df_2, how="outer", on="Date_time")
df_merge = pd.merge(df_merge, df_3, how="outer", on="Date_time")
df_merge = pd.merge(df_merge, df_4, how="outer", on="Date_time")
df_merge = pd.merge(df_merge, df_5, how="outer", on="Date_time")

df_merge = df_merge.sort_values(by='Date_time')
df_merge
```

```python
df_merge.columns
```

```python
df_merge["Mean_PM25"] = df_merge[["PM25_x","PM25_y", "PM25"]].mean(skipna=True,axis=1)
df_merge["Mean_PM10"] = df_merge[["PM10_x","PM10_y", "PM10"]].mean(skipna=True,axis=1)
df_merge["Mean_NO2"] = df_merge[["NO2_x","NO2_y", "NO2"]].mean(skipna=True,axis=1)
df_merge["Mean_O3"] = df_merge[["O3_x","O3_y", "O3"]].mean(skipna=True,axis=1)
df_merge["Mean_SO2"] = df_merge[["SO2_x","SO2_y", "SO2"]].mean(skipna=True,axis=1)

df_merge = df_merge.drop(columns=[
    "PM25_x","PM25_y", "PM25",
    "PM10_x","PM10_y", "PM10",
    "NO2_x","NO2_y", "NO2",
    "O3_x","O3_y", "O3",
    "SO2_x","SO2_y", "SO2"])
```

```python
df_merge = df_merge.rename(columns={
    "Mean_PM25":"PM25",
    "Mean_PM10":"PM10",
    "Mean_NO2":"NO2",
    "Mean_O3":"O3",
    "Mean_SO2":"SO2"})
```

```python
df_merge
```

```python
df = preprocess(df_merge)
```

```python
df
```

```python
df.to_csv('../../data/pollution/5_Clusters/cluster1_Ouest.csv', index=True)
```
