```python
#Supprimer l'horaire afin de garder uniquement la date
data['Date_time'] = data['Date_time'].astype(str).str[0:10]
data['Date_time']
```

```python
#Créer une ouvelle colonne contenant la moyenne PM10 et PM25
data_new = data.groupby(by='Date_time')[["PM10"]].mean()
data_new["PM25"] = data.groupby(by='Date_time')[["PM25"]].mean()
data_new
```

```python
#Créer une ouvelle colonne "NO2" et "O2" contenant la valeur max pour chaque journée
data_new["NO2"] = data.groupby(by='Date_time')[["NO2"]].max()
data_new
```
